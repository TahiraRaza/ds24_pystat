### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure: The overall window/container for all plot elements.
Axes: The area within a figure where data is plotted.
Axis: The x or y line that defines limits and ticks in an axes.
Artist: Any visual element (lines, markers, text) drawn in the figure.



### When plotting in Matplotlib, what is the expected input data type?

Plotting functions in Matplotlib expect numpy.array or numpy.ma.masked_array as input, or objects that can be passed to numpy.asarray


### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
import matplotlib.pyplot as plt
# pyplot approach

# Specify the function
x = np.linspace(-4, 4, 100)  # shows 100 points from -4 to 4
y = x**2

# Create the plot
plt.plot(x, y)

# Create the title and labels
plt.title('y = x^2')
plt.xlabel('x label')
plt.ylabel('y label')

# Show plot
plt.show()  
```

```python
# object-oriented approach

#Specify the function
x = np.linspace(-4, 4, 100)  # Shows 100 points from -4 to 4
y = x**2

# Create a figure and an axes using plt.subplots
fig, ax = plt.subplots()

# Create the plot
ax.plot(x, y)

# Create the title and labels
ax.set_title('y = x^2')
ax.set_xlabel('x values')
ax.set_ylabel('y values')

# Show plot
plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Create a figure and an axes using plt.subplots
fig, ax = plt.subplots()

# Create the scatter plot
ax.scatter(random_data_x, random_data_y)

# Create the title and labels
ax.set_title('y = x^2')
ax.set_xlabel('Random X data')
ax.set_ylabel('Randon Y data')

# Show plot
plt.show()

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Create a figure and an axes using plt.subplots
fig, ax = plt.subplots()

# Create the bar plot
ax.bar(names, values)

# Create the title and labels
ax.set_title('Fruit_data')
ax.set_xlabel('Name of the fruits')
ax.set_ylabel('Quantity')

# Show plot
plt.show()

```
